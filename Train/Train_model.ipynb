```python
import pickle
from TrainHelper import *
```

```python
## Config
config_species= 'gibbon_params'
model_name = 'CNN'
number_of_run = 10
train = TrainHelper()

def load_spectrograms_from_pickle(pickle_file):
    with open(pickle_file, 'rb') as f:
        spectrogram_array = pickle.load(f)
    return spectrogram_array
```

## Baseline

```python
method_0 = 'original'
compression = 0

X_S = './Out_Data/Baseline/X_raw.pkl'
Y_S = './Out_Data/Baseline/Y_S.pkl'
X = load_spectrograms_from_pickle(X_S)
Y = load_spectrograms_from_pickle(Y_S)

train.run_training(X,Y, config_species, model_name, method_0, compression, number_of_run)
```
